The following code loads the dataset and plots the first 25 imagesfrom the training set in 5 by 5 square

```
# import the function load_data 
from keras.datasets.mnist import load_data
from matplotlib import pyplot
# load train and test data
(trainX, trainy), (testX, testy) = load_data()
# Summarize the shape of dataset
print('Train', trainX.shape, trainy.shape)
print('Test', testX.shape, testy.shape)
# plot the first 25 images of training dataset
for i in range(25):
  # define subplot
  pyplot.subplot(5, 5, i + 1)
  # turn off axis
  pyplot.axis('off')
  # plot raw pixel data
  pyplot.imshow(trainX[i], cmap = 'gray_r') # cmap = gray_r is used to plot black digits on white background 
pyplot.show()
```
